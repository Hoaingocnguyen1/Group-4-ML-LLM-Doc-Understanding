```python


import openai

openai.api_key = ''

def generate_embedding(text):
    response = openai.Embedding.create(input=text, model='text-embedding-ada-002')
    return response.data[0].embedding

text = "Hello, how are you?"
embedding = generate_embedding(text)
print(embedding)
```
